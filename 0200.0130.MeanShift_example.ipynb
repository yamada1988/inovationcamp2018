# simple example of MeanShift

```python
from __future__ import print_function
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

"""
1. data preparation
"""
def load_data():
    """
    read data from     Maeda_carbon8_descriptor.csv
    
    @return : 2D p.array
    """
    df = pd.read_csv("Maeda_carbon8_descriptor.csv",index_col = [0,1])
    """
    use columns ["a0.5_rp1.5","a0.5_rp2.5"]
    """
    X = df[["a0.5_rp1.5","a0.5_rp2.5"]].as_matrix()
    print(X)
    """
    make 2D array

    plot raw data
    """
    plt.plot(X[:,0],X[:,1],".")
    plt.show()
    return X
X = load_data()
```

```python
"""

3. data analysis

"""
import sklearn.cluster
def analyze_X(X,n_clusters):
    """
    clustering
    """
    
    clf =  sklearn.cluster.MeanShift(seeds=)
    clf.fit(X)
    distortion = clf.inertia_
    cluster_centers = clf.cluster_centers_
    
    y = clf.predict(X) 

#    print(y.shape)
#    print (y)
#    print(distortion)
    return y, distortion, cluster_centers

n_clusters = 3
y, distortion, cluster_centers = analyze_X(X,n_clusters)
```

```python
y
import collections
def howmany(y):
    """
    check the number of labels 
    """
    counter = collections.Counter(y)
    print(counter)
    
howmany(y)
```

```python
"""
4. visualization
"""
import matplotlib.cm as cm
import numpy as np

def plot_result(X,y,n_clusters, cluster_centers):
    """
    plot them with different colors corresponding to cluster ids
    """
    
    for i in range(n_clusters):
        xx = []
        for x,y1 in zip(X,y):
            if y1 == i:
                xx.append(x)
        #print(xx)
        xx = np.array(xx)
        plt.scatter(xx[:,0],xx[:,1],color=cm.hsv(i/float(n_clusters)),marker='o',label=str(i),
                    edgecolors="black",s=100,)
    plt.scatter(cluster_centers[:,0],   # km.cluster_centers_には各クラスターのセントロイドの座標が入っている
                cluster_centers[:,1],
                s=500,
                marker='*',
                c='silver',
                label='centroids')
    plt.legend(loc='upper right', bbox_to_anchor=(1.35,1.0))
    plt.show()
plot_result(X,y,n_clusters, cluster_centers)
```

## エルボー法

```python
distortions = []
n_c=50

for i in range(1,n_c):
    y,distortion,cluster_centers = analyze_X(X,i)
    distortions.append(distortion)

plt.plot(range(1,n_c),distortions,marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.yscale("log")
plt.show()
```
